# Synthetic Handwriting With GAN

The goal of this research project is to use a Generative adversarial Neural network (GAN) to create synthetic handwritten letters. The dataset used for training the GAN is the [EMNIST Balanced](https://www.nist.gov/itl/products-and-services/emnist-dataset) which contains different classes 10 for the numbers 0 to 9 and 37 for the letters (some upper and lower case letter classes were merged because of simmilarity).
